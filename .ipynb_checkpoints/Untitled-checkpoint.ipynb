```python
from datetime import datetime
from datetime import timedelta
import numpy as np
import pandas as pd

data = pd.read_csv('useful_no_solar/run_household_{}/power2016Household.csv'.format(1), usecols=[2])
data.rename({'Power consumed': 'PowerConsumed'}, inplace=True)
base = datetime(2016, 1, 1)
dt = [base + timedelta(minutes=x) for x in range(len(data))]
data['date'] = dt
data = data.set_index('date')
```
